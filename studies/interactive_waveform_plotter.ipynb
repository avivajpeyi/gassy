```python
import ipywidgets as widgets
from gassy.two_body import WaveformGenerator
from IPython.display import display


def make_plot(m=1, M=1, r=-1, num_periods=10, drag_coeff=1e-5):
    kwgs = dict(m=m, M=M,  r=r, num_periods=num_periods, drag_coeff=drag_coeff)
    return WaveformGenerator(**kwgs).plot(distance=1000)



make_plot()
```

```python
def make_interactive_plot(r, drag_coeff):
    make_plot(r=r, drag_coeff=drag_coeff)

r_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=-1, continuous_update=False)
drag_slider = widgets.FloatLogSlider(min=-3, max=-1, value=-2,continuous_update=False)
l = widgets.Layout(min_height='50px', width='100px')
w=widgets.interactive(make_interactive_plot,r=r_slider, drag_coeff=drag_slider, layout=l)
w
```
