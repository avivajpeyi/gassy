```python
import ipywidgets as widgets
from gassy.two_body import WaveformGenerator
from IPython.display import display


def make_plot(m=1, M=1e4, init_x=-1, init_vy=0.025, num_periods=3, drag_coeff=1e-5):
    kwgs = dict(m=m, M=M, init_x=init_x, init_vy=init_vy, num_periods=num_periods, drag_coeff=drag_coeff)
    return WaveformGenerator(**kwgs).plot(distance=1000)



make_plot()
```

```python
def make_interactive_plot(init_x, init_vy, drag_coeff):
    make_plot(init_x=init_x, init_vy=init_vy, drag_coeff=drag_coeff)

init_x_slider = widgets.FloatSlider(min=-5, max=5, step=0.1, value=-1, continuous_update=False)
init_vy_slider = widgets.FloatSlider(min=0, max=0.03, step=0.001, value=0.025, continuous_update=False)
drag_slider = widgets.FloatLogSlider(min=-7, max=-4, value=-7,continuous_update=False)
l = widgets.Layout(min_height='50px', width='100px')
w=widgets.interactive(make_interactive_plot,init_x=init_x_slider,init_vy=init_vy_slider, drag_coeff=drag_slider, layout=l)


w
```
